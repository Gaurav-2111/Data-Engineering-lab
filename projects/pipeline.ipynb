```python
import requests
import json
import time

def extract():
    url = 'https://jsonplaceholder.typicode.com/todos'
    try:
        session = requests.Session()
        response = session.get(url)


        # api succesfully connected
        if response.status_code == 200:
          print(response.status_code)
          print(response.headers)
          content_type = response.headers.get('content_type')
          # checking if the data returning is json
          if 'application/json' in content_type:
            data = response.json()

          else:
            print("data returned is not json")

        elif response.status_code == 204:
          print("No record found")
          # no point in retrying as data does not exist

        else:
            response.raise_for_status()

    # STATUS CODE ERROR
    except requests.exceptions.HTTPError as e:
      # TOO MANY REQUESTS
      if response.status_code == 429:
        print("too many requests error : ",response.status_code)
        # DELAY BY 1 SEC
        time.sleep(1)

      # authentication failure
      elif response.status_code == 401:
        print("authentication problem , status_code : ",response.status_code)
        # login again

      # unauthorized access
      elif response.status_code == 403:
        print("Access denied you do not have the permission to acess the resource : ",response.status_code)
        # log the error to tell the dev no point in retrying as you don not have the access

      # missing resourcer
      elif response.status_code == 404:
        print("resource you are looking is missing : ",response.status_code) # no point in retrying


      # internal code crashed
      elif response.status_code == 500:
        print("internal server error , status_code : ",response.status_code)

      # SERVER overloaded or in maintenance mode
      elif response.status_code == 503:
        print("service unavailable , status_code : ",response.status_code)

      # Gateway can not talk to backend
      elif response.status_code == 502:
        print("bad gateway , status_code : ",response.status_code)

      # something else happened
      else:
        print("something went wrong , status_code : ",response.status_code)


    # Network related issues
    except requests.exceptions.ConnectionError as conn_err:
        print(f'Connection error occurred: {conn_err}')

    # Timeout server took too long to reply to our api
    except requests.exception.Timeout as timeout_err:
        print('server took too long to reply')



```
